```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
        
    def insert(self, value):
        if self.head == None:
            self.head = Node(value)
        else:
            new_node = Node(value)
            temp = self.head
            while temp.next:
                temp = temp.next
            temp.next = new_node
    
    def printList(self):
        
        if self.head == None:
            return 0
        else:
            temp = self.head
            while temp.next:
                print(temp.data,end=" -> ")
                temp = temp.next
            print(temp.data)
```

```python
def countList(head):
        count = 1
        if head == None:
            return 0
        else:
            temp = head
            while temp.next:
                count += 1
                temp = temp.next
            return count
```

```python
def findMid(head):
    slow = head
    fast = head
    
    while fast.next:
        slow = slow.next
        if fast.next.next:
            fast = fast.next.next
        else:
            fast = fast.next
    return slow.data
```

```python
def getNthFromEnd(head, n):
    
    main,ref = head,head
    count = 1
    
    while count != n:
        if ref.next == None:
            return -1
        ref = ref.next
        count += 1
    
    while ref.next:
        main = main.next
        ref = ref.next
    
    return main.data
    
```

```python
def rotateList(head, k):
    temp = head
    count = 1
    
    while count != k:
        temp = temp.next
        count += 1
    
    kth_node = temp
    
    while temp.next:
        temp = temp.next
    
    temp.next = head
    head = kth_node.next
    kth_node.next = None
    
    return head 
```

```python
def reverseList(head):
    
    if head is None:
        return None
    
    prev, next_ = None, None
    curr = head
    
    while curr:
        
        next_ = curr.next
        curr.next = prev
        
        prev = curr
        curr = next_
    
    return prev
```

```python
def reverseList(head): # Using Recursion
    
    if head.next == None:
        return head
    node = reverseList(head.next)
    head.next.next = head
    head.next = None
    return node
```

```python
def reverseInGroups(head, k):
    next_, prev = None, None
    curr = head
    count = 0
    
    while count != k and curr:
        next_ = curr.next 
        curr.next = prev
        prev = curr
        curr = next_
        count += 1
    
    if next_:
        head.next = reverseInGroups(next_, k)
    
    return prev
    
```

```python
def detectLoop(head):
    slow = fast = head
    
    while slow and fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        
        if fast == slow:
            return 1
    return 0
```

```python
def detectAndRemoveLoop(head):
    slow = head.next
    fast = head.next.next
    
    while fast is not None:
        if (fast.next is None) or (slow == fast):
            break
        
        slow = slow.next
        fast = fast.next.next
        
    if slow == fast:
        slow = head
        
        while slow.next is not fast.next:
            slow = slow.next 
            fast = fast.next

        fast.next = None
        return 1
    return 0
    
```

```python
def createLoop(head, val):
        
    temp = head
    for i in range(1,val):
        temp = temp.next
    
    loop = temp
    
    while temp.next:
        temp = temp.next
    
    temp.next = loop
```

# Intersection of two Linked List

```python
intersect_a = LinkedList()
intersect_b = LinkedList()
```

```python
intersect_a.head.next = second
second.next = third
third.next = four

intersect_b.head.next = third
```

Solution 1: Use 2 nested for loops. The outer loop will be for each node of the 1st list and inner loop will be for 2nd list. In the inner loop, check if any of nodes of the 2nd list is same as the current node of the first linked list. The time complexity of this method will be O(M * N) where m and n are the numbers of nodes in two lists.

```python
def intersectionPoint(head_a, head_b):
    temp_a = head_a
    
    while temp_a:
        temp_b = head_b
        while temp_b:
            if temp_a == temp_b:
                return temp_b.data
            temp_b = temp_b.next
        temp_a = temp_a.next
    return -1
```

```python
intersectionPoint(intersect_a.head, intersect_b.head)
```

Solution 2: Traverse the first linked list and store the addresses of visited nodes in a hash. Now traverse the second linked list and if you see an address that already exists in the hash then return the intersecting node.

```python
def intersectPoint(head_a, head_b):
    hash_ = {}
    
    temp_a = head_a
    while temp_a:
        hash_[temp_a] = 1
        temp_a = temp_a.next
    
    temp_b = head_b
    while temp_b:
        if temp_b in hash_:
            return temp_b.data
        temp_b = temp_b.next
    return -1
```

```python
intersectPoint(intersect_a.head, intersect_b.head)
```


Method 3(Using difference of node counts)

Get count of the nodes in the first list, let count be c1.
Get count of the nodes in the second list, let count be c2.
Get the difference of counts d = abs(c1  c2)
Now traverse the bigger list from the first node till d nodes so that from here onwards both the lists have equal no of nodes.
Then we can traverse both the lists in parallel till we come across a common node. (Note that getting a common node is done by comparing the address of the nodes)

```python
def intersetPoint(head_a, head_b):
    
    c1 = c2 = 1
    temp_a = head_a
    temp_b = head_b
    
    # Get count of the nodes in the first list,
    # let count be c1
    while temp_a.next:
        temp_a = temp_a.next
        c1 += 1
    
    # Get count of the nodes in the second list,
    # let count be c2
    while temp_b.next:
        temp_b = temp_b.next
        c2 += 1
    
    d = abs(c1-c2)
    
    curr_1 = head_a if c1 > c2 else head_b
    curr_2 = head_b if curr_1 == head_a else head_a
    
    # Now traverse the bigger list from the first node till d nodes 
    # so that from here onwards both the lists have equal no of nodes
    while d:
        curr_1 = curr_1.next
        d -= 1
    
    # traverse both the lists in parallel till we come across a common node
    while curr_1 != curr_2:
        curr_1 = curr_1.next
        curr_2 = curr_2.next
    
    return curr_1.data if curr_1 == curr_2 else -1
```

```python
intersetPoint(intersect_a.head, intersect_b.head)
```

# Merge two Sorted Linked List

```python
def merge(head_a, head_b):
    
    if head_a is None:
        return head_b
    
    if head_b is None:
        return head_a
    
    if head_a.data <= head_b.data:
        temp = head_a
        temp.next = merge(head_a.next, head_b)
    else:
        temp = head_b
        temp.next = merge(head_a, head_b.next)
    
    return temp
```

```python
l1 = [2,3,20]
l2 = [5,10,15,40]
```

```python
merge_l2 = LinkedList()
for i in l2:
    merge_l2.insert(i)
merge_l2.printList()
```

```python
merge_ = LinkedList()
merge_.head = merge(merge_l1.head, merge_l2.head)
merge_.printList()
```

# Stack using Linked List

```python
class MyStack:
    
    def __init__(self):
        self.head = None
    
    def push(self, val):
        new_node = Node(val)
        
        if self.head is None:
            self.head = new_node
        else:
            new_node.next = self.head
            self.head = new_node
    
    def peek(self):
        
        if self.head is None:
            return None
        else:
            return self.head.data
    
    def pop(self):
        
        if self.head is None:
            return 'Stack is underflow'
        
        else:
            temp = self.head
            self.head = self.head.next
            print('{} is deleted from the stack'.format(temp.data))
            del(temp)
            return
    
    def printStack(self):
        
        if self.head is None:
            return 'Stack is Empty'
        else:
            print(self.head.data,end="<-- Head\n")
            temp = self.head.next
            while temp:
                print(temp.data)
                temp = temp.next
               
    
    
```

```python
stack = MyStack()
```

```python
stack.push(5)
stack.push(4)
stack.push(3)
stack.push(2)
stack.push(1)
```

```python
stack.printStack()
```

```python
stack.pop()
```

```python
stack.printStack()
```

```python
stack.peek()
```

# Queue With Linked List

```python
class Queue:
    
    def __init__(self):
        self.front = self.rear = None
    
    def isEmpty(self):
        return self.front == None
        
    def EnQueue(self, item):
        new_node = Node(item)
        
        if self.rear == None:
            self.front = new_node
            self.rear = new_node
            return 
        
        else:
            self.rear.next = new_node
            self.rear = new_node
            return
    
    def DeQueue(self):
        
        if self.front == None:
            return -1
        else:
            temp = self.front
            self.front = self.front.next
            
            if self.front is None:
                self.rear = None
            
            return temp.data
            
        
```

```python
q = Queue() 
q.EnQueue(10) 
q.EnQueue(20) 
q.DeQueue() 
q.DeQueue() 
q.EnQueue(30) 
q.EnQueue(40) 
q.EnQueue(50) 

print("Dequeued item is {}".format(q.DeQueue()))
```

# Sort a linked list of 0s, 1s and 2s

Following steps can be used to sort the given linked list.

Traverse the list and count the number of 0s, 1s and 2s. Let the counts be n1, n2 and n3 respectively.

Traverse the list again, fill the first n1 nodes with 0, then n2 nodes with 1 and finally n3 nodes with 2.

```python
def segregate(head):
    
    if head is None or head.next is None:
        return head
        
    
    zeroD = oneD = twoD = Node(0)
    
    zero = zeroD
    one = oneD
    two = twoD
    
    temp = head
    while temp:
        
        if temp.data == 0:
            zero.next = temp
            zero = zero.next
            temp = temp.next
        
        elif temp.data == 1:
            one.next = temp
            one = one.next
            temp = temp.next
        
        elif temp.data == 2:
            two.next = temp
            two = two.next
            temp = temp.next
        
    
    head = zeroD.next
    zero.next = oneD.next
    one.next = twoD.next
    
    del(zeroD)
    del(oneD)
    del(twoD)
    
    return head
```

```python
segregate_list = LinkedList()
lis = [1, 2, 2, 1, 2, 0, 2, 2]
for i in lis:
    segregate_list.insert(i)
segregate_list.printList()
```

```python
segregate(segregate_list.head)
```

# Pairwise Swap 

```python
def pairwiseSwap(head):
    
    new_head = ptr_one_prev = None
    
    while head and head.next:
        ptr_one = head
        ptr_two = head.next   
        ptr_two_next = ptr_two.next
        
        ptr_one.next = ptr_two_next
        
        if ptr_one_prev is not None:
            ptr_one_prev.next = ptr_two
        else:
            new_head = ptr_two
        
        ptr_two.next = head
        ptr_one_prev = ptr_one
        head = ptr_two_next
    
    return new_head 
```

```python
def pairwiseSwap(head):
    
    if head is None or head.next is None:
        return head
    
    ptr_one = head
    ptr_two = head.next
    ptr_two_next = ptr_two.next
    ptr_two.next = head
    new_head = ptr_two
    ptr_one.next = pairwiseSwap(ptr_two_next)
    return new_head
```

```python
pair = LinkedList()
pair_ele = [1,2,3,4,5,6]
for i in pair_ele:
    pair.insert(i)
pair.printList()
```

```python
result = (pairwiseSwap(pair.head))
printList(result)
```

```python
def isPalindrome(head):
    
    s = []
    temp = head
    
    while temp:
        s.append(temp.data)
        temp = temp.next
        
    if len(s) == 0:
        return False
    
    curr = head
    while curr:
        x = s.pop()
        if x != curr.data:
            return False
        
        curr = curr.next
    
    if len(s):
        return False
    else:
        return True
```

```python
palindrome = LinkedList()
palindrome_ele = [1,1,1,3]
for i in palindrome_ele:
    palindrome.insert(i)
palindrome.printList()
```
