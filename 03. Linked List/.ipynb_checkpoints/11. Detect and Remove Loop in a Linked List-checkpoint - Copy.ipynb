## Detect Loop in a Linked List

checks whether a given Linked List contains loop and if loop is present then removes the loop and returns true. If the list doesnâ€™t contain loop then it returns false. Below diagram shows a linked list with a loop. detectAndRemoveLoop() must change the below list to 1->2->3->4->5->NULL.

<img src="https://www.geeksforgeeks.org/wp-content/uploads/2009/04/Linked-List-Loop.gif" >

```python
class Node:
    def __init__(self ,data):
        self.data = data
        self.next = None
```

```python
def insert(head ,node):
    if head == None:
        head = node
        return head
    temp = head
    while temp.next:
        temp = temp.next
    temp.next = node
    return head
```

```python
def printList(head):
    if head == None:
        return 0
    else:
        temp = head
        while temp.next:
            print(temp.data,end=" -> ")
            temp = temp.next
        print(temp.data)
```

```python
head = None
lis_5_a = [1, 2, 3, 4 ,5, 6]
for i in lis_5_a:
    head = insert(head ,Node(i))
printList(head)
```

```python
def insertLoop(head ,k = 3):
    temp = head
    while temp and k-1:
        temp = temp.next 
        k -= 1
    
    curr = temp
    
    while temp.next:
        temp = temp.next
    
    temp.next = curr
```

```python
insertLoop(head)
```

```python
printList(head)
```

```python
def detectLoop(head):
    slow = head
    fast = head
    
    while slow and fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        
        if slow == fast:
            return slow.data
    
    return None
```

```python
detectLoop(head)
```

```python
printList(head)
```

## Detect and Remove Loop from the Linked List

```python
printList(head)
```

```python
def detectAndRemove(head):
    
    slow = head
    fast = head
    
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        
        if slow == fast:
            return slow.data
    
    return -1
```

```python
detectAndRemove(head)
```

```python
printList(head)
```
